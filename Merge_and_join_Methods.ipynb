```python
import pandas as pd
```

```python

data1 = {
    "ID" : [10001,20002,30003,40004,50005],
    "Numbers": [10,20,30,40,50],
    "Letters":["A","B","C","D","E"]
}
```

```python
df1 = pd.DataFrame(data1)
```

```python
df1
```

```python
data2 = {
    "ID" : [10001,20002,30003,60006,70007],
    "Numbers": [10,20,30,40,60],
    "City":["Lucknow","Munnar","Chennai","Delhi","Jaipur"]
}
```

```python
df2 = pd.DataFrame(data2)
```

```python
df2
```

```python
merge_data = df1.merge(df2,how='inner')
```

```python
merge_data
```

```python
merge_data = df1.merge(df2,how='inner',on='ID')
```

```python
merge_data
```

```python
merge_data = df1.merge(df2,how='inner',on=['ID','Numbers'])
```

```python
merge_data
```

```python
merge_data = df1.merge(df2,how='outer')
```

```python
merge_data
```

```python
merge_data = df1.merge(df2,how='outer', on='ID')
```

```python
merge_data
```

```python
merge_data = df1.merge(df2,how='outer', on=['ID','Numbers'])
```

```python
merge_data
```

```python
merge_data3 = df1.merge(df2,how='outer')
```

```python
merge_data3
```

```python
merge_data3 = df1.merge(df2,how='outer', on='ID')
merge_data3
```

```python
merge_data3 = df1.merge(df2,how='left')
merge_data3
```

```python
merge_data3 = df1.merge(df2,how='right')
merge_data3
```

```python
cross_merge = df1.merge(df2,how='cross')
cross_merge
```
